```python
data_path2 = "C:/Users/GCU/Lending_club/Data_Analysis_lending-club/tablegan/data/defalut_original/defalut_real.csv"

data2 = pd.read_csv(data_path2, low_memory=False)

data2.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# ‚úÖ Î≤îÏ£ºÌòï(Object) Ïª¨Îüº ÏÑ†ÌÉù
object_features = data2.select_dtypes(include=['object']).columns

# ‚úÖ Í∞Å Î≤îÏ£ºÌòï Ïª¨ÎüºÏóê ÎåÄÌï¥ Label Encoding Ï†ÅÏö©
encoders = {}  # Í∞Å Ïª¨ÎüºÏùò LabelEncoder Ï†ÄÏû•Ïö©
encoded_data = data2.copy()  # ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ Ïú†ÏßÄ

for col in object_features:
    encoder = LabelEncoder()
    encoded_data[col] = encoder.fit_transform(data2[col])  # Í∞Å Ïª¨ÎüºÏùÑ Í∞úÎ≥ÑÏ†ÅÏúºÎ°ú Î≥ÄÌôò
    encoders[col] = encoder  # Ïù∏ÏΩîÎçî Ï†ÄÏû•

# ‚úÖ Í∞Å Ïª¨ÎüºÏùò ÌÅ¥ÎûòÏä§ ÌôïÏù∏
for col, encoder in encoders.items():
    print(f"üîπ {col}: {encoder.classes_}")
```
