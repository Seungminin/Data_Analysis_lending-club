```python
data_path2 = "C:/Users/GCU/Lending_club/Data_Analysis_lending-club/tablegan/data/defalut_original/defalut_real.csv"

data2 = pd.read_csv(data_path2, low_memory=False)

data2.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# ✅ 범주형(Object) 컬럼 선택
object_features = data2.select_dtypes(include=['object']).columns

# ✅ 각 범주형 컬럼에 대해 Label Encoding 적용
encoders = {}  # 각 컬럼의 LabelEncoder 저장용
encoded_data = data2.copy()  # 원본 데이터 유지

for col in object_features:
    encoder = LabelEncoder()
    encoded_data[col] = encoder.fit_transform(data2[col])  # 각 컬럼을 개별적으로 변환
    encoders[col] = encoder  # 인코더 저장

# ✅ 각 컬럼의 클래스 확인
for col, encoder in encoders.items():
    print(f"🔹 {col}: {encoder.classes_}")
```

```python
encoded_data.head()
```

```python
##fake data 되돌리기
for col in object_features:
    fake_data[col] = encoders[col].inverse_transform(np.round(fake_data[col]).astype(int))
```
